```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# loading the data:
data=pd.read_csv("student_hours_marks.csv")
data.head
```

```python
# loss function:
def loss_function(w,b,points):
    total_error=0
    n=len(points)
    for i in range(len(points)):
        x=points.iloc[i].Hours
        y=points.iloc[i].Marks
        total_error+=((w*x+b)-y)**2
    return total_error/(2*n)
```

```python
# gradient descent:
def gradient_descent(w_now,b_now,points,l):
    w_gradient=0
    b_gradient=0
    n=len(points)
    for i in range(n):
        x=data.iloc[i].Hours
        y=data.iloc[i].Marks
        w_gradient+=(1/n)*((w_now*x)-y)*x
        b_gradient+=(1/n)*((w_now*x)-y)
    w=w_now-l*w_gradient
    b=b_now-l*b_gradient
    return w,b
```

```python
# initializing gradient parameters:
w=0
b=0
l=0.0001
epoch=1000
```

```python
# calling gradient descent for epochs times:
for i in range(epoch):
    w,b=gradient_descent(w,b,data,l)
print(w,b)
```

```python
# find the error:
print("Loss:")
print(loss_function(w,b,data))
```

```python
# ploting:
plt.scatter(data.Hours,data.Marks,color="Green")
plt.plot(list(range(0,10)),[w*x +b for x in range(0,10)],color="red")
```
